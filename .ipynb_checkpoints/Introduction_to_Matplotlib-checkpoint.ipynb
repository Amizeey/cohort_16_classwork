# üìä Introduction to Matplotlib

## What is Matplotlib?

**Matplotlib** is a Python library used for **data visualization**.
It allows data scientists to **turn numbers into charts** so patterns, trends, and insights are easy to see.

In Data Science, we rarely understand data just by looking at tables.
**Visualization helps us ask better questions and make better decisions.**

## Why Matplotlib is Important in Data Science

Matplotlib helps us:

* Understand trends over time
* Compare values between groups
* Detect outliers and unusual patterns
* Communicate insights clearly to others

It is often the **first visualization library** data scientists learn.

## Installing and Importing Matplotlib

Matplotlib is usually imported using `pyplot`, which behaves like a drawing canvas.

```python
import matplotlib.pyplot as plt
```

`plt` is a convention used by most data scientists.

---

## The Core Idea Behind Matplotlib

Matplotlib works using **Figures and Axes**.

* **Figure** ‚Üí The entire window or page
* **Axes** ‚Üí The actual plot area (where data is drawn)

You don‚Äôt always create them manually, but understanding this helps later.

## Basic Plot Types

### 1Ô∏è‚É£ Line Plot

Used to show **trends over time**.

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

**Use cases:**

* Sales over months
* Website visits per day
* Stock price movement

### 2Ô∏è‚É£ Bar Chart

Used to **compare categories**.

```python
products = ['A', 'B', 'C', 'D']
sales = [100, 150, 80, 100]

plt.bar(products, sales)
plt.show()
```

**Use cases:**

* Product comparison
* Performance by category
* Count of observations

### 3Ô∏è‚É£ Scatter Plot

Used to show **relationships between two variables**.

```python
hours = [1, 2, 3, 4, 5]
scores = [50, 55, 65, 70, 85]

plt.scatter(hours, scores)
plt.show()
```

**Use cases:**

* Study hours vs exam score
* Age vs income
* Any correlation analysis

**Use cases:**

* Identifying extreme values (anomalies)
* Comparing spread between different groups
* Statistical summaries

```python
values = [10, 20, 25, 30, 35, 40, 45, 100]  # 100 is an outlier

plt.boxplot(values)
plt.show()
```

### 6Ô∏è‚É£ Box Plot

Used to detect **outliers** and see data spread.

**Use cases:**

* Market share analysis
* Budget allocation
* Composition of a dataset

```python
categories = ['Rent', 'Food', 'Transport', 'Savings']
expenses = [40, 30, 10, 20]

plt.pie(expenses, labels=categories, autopct='%1.1f%%')
plt.show()
```

### 5Ô∏è‚É£ Pie Chart

Used to show **parts of a whole** (percentages).

**Use cases:**

* Age groups in a survey
* Distribution of exam scores
* Checking if data is normal (Bell curve)

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7]

plt.hist(data, bins=2, edgecolor='black')
plt.show()
```

### 4Ô∏è‚É£ Histogram

Used to see **distribution of data** (frequency).

## Adding Labels and Titles (Very Important)

Charts must always be **clear and readable**.

```python
plt.plot(x, y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Daily Sales Trend")
plt.show()
```

This helps:

* You understand the chart
* Others understand your analysis

## Customizing Plots (Basic Level)

### Change Line Style and Markers

```python
plt.plot(x, y, marker='o', linestyle='-.')
plt.show()
```

### Add Grid

```python
plt.plot(x, y)
plt.grid(True)
plt.show()
```

Grids help make values easier to read.

## Multiple Plots on the Same Chart

```python
y2 = [8, 18, 28, 35, 40]

plt.plot(x, y, label='Colab in 2018')
plt.plot(x, y2, label='Colab in 2025')
plt.legend()
plt.show()
```

Used for **comparison**.
