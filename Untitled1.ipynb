```python
# importing all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the dataset
df = pd.read_csv("social_media_viral_content_dataset.csv")
```

```python
# viewing the first few observations
df.head()
```

```python
df['post_datetime'] = pd.to_datetime(df['post_datetime'])
df['month_name'] = df['post_datetime'].dt.month_name()
df.head(20)
```

```python
df.info()
```

```python
# viewing the size of the dataset
df.shape
```

```python
# Getting the summary statistics of the data
df.describe()
```

### Task 1: Data Filtering

```python
# filtering the dataset based on number of likes 
viral_posts = df[df["views"] > 4000000]
viral_posts.head()
```

```python
viral_posts.shape
```

```python
viral_posts.describe()
```

### Task 2: Group Analysis

```python
engagement = viral_posts.groupby("content_type")["views"].agg(["mean", "count"])
engagement
```

```python
count_array = np.array(engagement["count"])
print ("Count Array:", count_array)
```

### Task 3: Comparative Visualization

```python
plt.hist(viral_posts["views"], bins= 5, edgecolor= "white")
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.title('Distribution of Views')
plt.show()
```

```python
plt.scatter(data= viral_posts.head(100), x = "likes", y= "views")
plt.xlabel("likes")
plt.ylabel("views")
plt.title("Likes vs Views (Top 100 Viral Posts)")
plt.show()
```

### Task 4: Trend & Distribution Analysis

```python
monthly_views = viral_posts.groupby('month_name')['views'].sum().reset_index()
monthly_views
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_views['month_name'], monthly_views['views'], marker='o')
plt.xlabel('Months')
plt.ylabel('Total Views')
plt.title('Monthly Views Trend')
plt.grid(True)

# Optional: format y-axis in millions
plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
)

plt.show()
```
